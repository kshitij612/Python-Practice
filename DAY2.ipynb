```python
import logging
logging.basicConfig(
    filename = "logs",
    level = logging.INFO,
    format="%(asctime)s -%(levelname)s - %(message)s"
)


def vowel(str: str) -> int:
    '''Count the number of vowels.'''
    logging.info("Counting vowels in a string")
    count = 0
    for i in str:
        if(i in 'aeiouAEIOU'):
            count+=1
    logging.info(f"Vowel count calculated: {count}")
    return count

vowel('aeiouAEIOU')
```

```python
import logging
logging.basicConfig(
    filename = "frequency.log",
    level = logging.INFO,
    format="%(asctime)s -%(levelname)s - %(message)s"
)
def charfreq(str:str) -> dict:
    '''Counting the frequency of cahracters in a string.'''
    logging.info("Counting the frequency of charcters.")
    dict = {}
    for i in str:
        if(i in dict):
            dict[i] += 1
        else:
            dict[i] = 1
    logging.info(f"Frequency Counted: {dict}")
    return dict
charfreq("")
```
