# POLYMORPHISM

```python
#python doesnt support true method overloading
#we achieve it using default args and variable length args
class math:
    def add(self, a = 0, b = 0, c= 0):
        return a*b*c
obj = math()
obj.add(3,4,6)
```

```python
#monkey patching: phenomena of assigning a fucntion reference 

def sam():
    print("Polymorphism")
ref = sam

def sam(a,b):
    print(a+b)

sam(2,3)
ref()
```

```python
class Veichle:
    def __init__(self, veichle_id, brand):
        self.veichle_id = veichle_id
        self.brand = brand
    
    def calculate_rent():
        pass
    
class Car(Veichle):
    def __init__(self, veichle_id, brand, rent_per_day, days):
        super().__init__(veichle_id, brand)
        self.rent_per_day = rent_per_day
        self.days = days

    def calculate_rent(self):
        return self.days * self.rent_per_day

class Bike(Veichle):
    def __init__(self, veichle_id, brand, rent_per_hour, hour):
        super().__init__(veichle_id, brand)
        self.rent_per_hour = rent_per_hour
        self.hour = hour
    
    def calculate_rent(self):
        return self.hour * self.rent_per_hour

class Truck(Veichle):
    def __init__(self, veichle_id, brand, rent_per_km, distCovered):
        super().__init__(veichle_id, brand)
        self.rent_per_km = rent_per_km
        self.distCovered = distCovered
    
    def calculate_rent(self):
        return self.rent_per_km * self.distCovered
    
car = Car(123,"Honda", 100,10)
print(car.calculate_rent())

bike = Bike(234,"Ford", 5000, 5)
print(bike.calculate_rent())

truck = Truck(547, "Tata", 200, 50)
print(truck.calculate_rent())
```
