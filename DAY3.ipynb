```python
class Bank:
    bank_name = "SBI"
    branch = "Haldia"
    ifsc = "SBIN00032"
    country = "India"

    #Constructor to initialize object members
    def __init__(self, name, age, phone, pan, balance):
        self.name = name
        self.age = age
        self.phone = phone
        self.pan = pan 
        self.balance = balance
    
    def withdraw(self, balance):
        if(self.balance > balance and balance > 0):
            self.balance = self.balance - balance
    def deposit(self, balance):
        self.balance = self.balance + balance

#Creating objects 
c1 = Bank("Aditya", 22, 9489489489, "LXHFL493R", 100)
c2 = Bank("Aman", 25, 2432948949, "LKBFN7488R", 80000)
c3 = Bank("Bumba", 29, 567832489, "KNKGH1542I", 12000)

print(c1.name, c1.age, c1.balance)
c1.withdraw(-500)

print(c1.name, c1.age, c1.balance)
c1.deposit(10000)

print(c1.name, c1.age, c1.balance)

print(c2.name, c2.age, c2.balance)
```

```python
class Bank:
    bank_name = "SBI"
    branch = "Haldia"
    ifsc = "SBIN00032"
    country = "India"

    #Constructor to initialize object members
    def __init__(self, name, age, phone, pan, balance):
        self.name = name
        self.age = age
        self.phone = phone
        self.pan = pan 
        self.balance = balance
    
    def display(self):
        print(self.name, self.phone, self.balance)
    def withdraw(self, balance):
        if(self.balance > balance and balance > 0):
            self.balance = self.balance - balance
    def deposit(self, balance):
        self.balance = self.balance + balance

    @classmethod    
    def branch_Change(cls, branch):
        cls.branch = branch

#Creating objects 
c1 = Bank("Aditya", 22, 9489489489, "LXHFL493R", 100)
c2 = Bank("Aman", 25, 2432948949, "LKBFN7488R", 80000)
c3 = Bank("Bumba", 29, 567832489, "KNKGH1542I", 12000)

print(c1.name, c1.age, c1.balance, c1.branch)

Bank.branch_Change("Kolkata")


print(c2.name, c2.age, c2.balance, c2.branch)
```
