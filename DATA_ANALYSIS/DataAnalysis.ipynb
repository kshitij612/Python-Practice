Merge and Concat

```python
data1 = {
    'Name':['A','B','C'],
    'age':[21,22,32],
    'salary': [10000,20000,30000]
    }

data2 = {
    'Name': ['A','BB','C',],
    'age':[42,54,32],
    'salary': [1000,25000,30000]
}
```

```python
new_df1 = pd.DataFrame(data1)
new_df2 = pd.DataFrame(data2)
```

```python
new_df1.head()
```

```python
new_df2.head()
```

```python
pd.concat([new_df1, new_df2], ignore_index=True, axis=0)
```

```python
new_df = pd.concat([new_df1, new_df2], ignore_index=True, axis=0)
new_df
```

```python
new_df.drop_duplicates(inplace=True)
new_df.head(7)
```
