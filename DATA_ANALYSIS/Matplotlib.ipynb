```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1, 50, 10)
y = np.random.randint(1,50,10)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color = 'red', linestyle = 'dotted')
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.plot(x,y, color = 'red', linestyle = 'dotted')
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y, color = 'red', linestyle = 'dotted')
plt.show()
```

```python
x= np.arange(1,10)
y = np.sin(x)
z = np.cos(x)

print(x,y,z, sep = '\n\n')
```

```python
plt.plot(x,y, label = 'SINQ', color = 'red')
plt.plot(x,z, label = 'COSQ')
plt.title('SINQ & COSQ')
plt.legend()
plt.show()
```

```python
plt.plot(x,y, label = 'SINQ', color = 'red')
plt.plot(x,z, label = 'COSQ')
plt.title('SINQ & COSQ')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y, label = 'SINQ', color = 'red')
plt.plot(x,z, label = 'COSQ')
plt.title('SINQ & COSQ')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.title('SINQ & COSQ')
plt.plot(x,y, label = 'SINQ', color = 'red')
plt.plot(x,z, label = 'COSQ')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.legend()

plt.subplot(1,2,2)
plt.title('SINQ & COSQ')
plt.plot(x,y, label = 'SINQ', color = 'green')
plt.plot(x,z, label = 'COSQ', color = 'red')
plt.xlabel('Data Points 2')
plt.ylabel('Associate Values')


plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,9))

plt.subplot(3,3,1)
plt.title('SINQ & COSQ')
plt.plot(x,y, label = 'SINQ', color = 'red')
plt.plot(x,z, label = 'COSQ')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.legend(loc = 'upper right')

plt.subplot(3,3,5)
plt.title('SINQ & COSQ')
plt.plot(x,y, label = 'SINQ', color = 'green')
plt.plot(x,z, label = 'COSQ', color = 'red')
plt.xlabel('Data Points 2')
plt.ylabel('Associate Values')
plt.legend(loc = 'upper right')
plt.subplots_adjust(wspace=0.8)

plt.subplot(3,3,9)
plt.title('SINQ & COSQ')
plt.plot(x,y, label = 'SINQ', color = 'blue')
plt.plot(x,z, label = 'COSQ', color = 'red')
plt.xlabel('Data Points 3')
plt.ylabel('Associate Values')

plt.legend()
plt.show()
```
