```python
class A:
    a = 10
    b = 20

    @classmethod
    def display(cls):
        print(cls.a, cls.b)

class B(A):
    c = 30
    def __init__(self,m):
        self.m = m

obj1 = A()
obj2 = B(300)
obj1.display()
```

```python
class Bank:
    bank_name = "SBI"
    branch = "Haldia"

    def __init__(self, name , phone, age):
        self.name = name
        self.phone = phone
        self.age = age
    def display(self):
        print(self.name, self.phone, self.age)
    
    @classmethod
    def disp(cls):
        print(cls.bank_name, cls.branch)


class Bank_Update(Bank):
    def __init__(self, name, phone, age, email, pan, aadhar):
        #Constructor chaining
        super().__init__(name, phone, age)
        self.email = email
        self.pan = pan
        self.aadhar = aadhar
    def show_details(self):
        #Method chaining
        super().display(self)
        print(self.email, self.pan, self.aadhar)

c1 = Bank_Update(
    "Kshitij", 
    "87854121220",
    "12",
    "kshitij@gmail.com",
    "LHGBI8768H",
    "554422169521"
)
Bank_Update.disp()
# Bank_Update.display(c1)
c1.display()
```

```python
#INHERITENCE

class Tenth:
    name = "Kshitij"
    loc = "Haldia"

    def __init__(self, fav_sub):
        self.fav_sub = fav_sub
    
        
class Twelth(Tenth):
    
    def __init__(self, fav_subject, stream, batch):
        super().__init__(fav_subject)
        self.stream = stream
        self.batch = batch
    def display(self):
        print(self.name, self.loc, self.fav_sub, self.stream, self.batch)

s1 = Twelth("Maths", "PCM", "2")
s1.display()
    
```

```python
#MULTI LEVEL INHERITENCE

class Tenth:
    name = "Kshitij"
    loc = "Haldia"

    def __init__(self, fav_sub):
        self.fav_sub = fav_sub
    
        
class Twelth(Tenth):
    
    def __init__(self, fav_subject, stream, batch):
        super().__init__(fav_subject)
        self.stream = stream
        self.batch = batch
    def display(self):
        print(self.name, self.loc, self.fav_sub, self.stream, self.batch, end = " ")

class bachelor(Twelth):
    def __init__(self, fav_subject, stream, batch, branch, uniName):
        super().__init__(fav_subject, stream, batch)
        self.branch = branch
        self.uniName = uniName

    def display(self):
        super().display()
        print(self.branch, self.uniName)
        

s1 = Twelth("Maths", "PCM", "2")
s1.display()

s2 = bachelor("Maths","PCM","1","CSE","HIT")
s2.display()
    
```

```python
#HIERARICHAL INHERITANCE

class A:
    a = 10
    b = 20
    def __init__(self, c, d):
        self.c = c 
        self.d = d

class B(A):
    a = 1000
    @classmethod
    def display(cls):
        print(cls.a)

class C(A):
    a = 2000

obj1 = B(1,2)
obj2 = C(10,20)
print(obj1.a, obj1.b, obj1.c, obj1.d)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, sId, rollNum):
        super().__init__(name, age)
        self.sId = sId
        self.rollNum = rollNum

# class Teacher(Person):
#     def __init__(self, name, age, tId, sub):
#         super().__init__(name, age)
#         self.tId = tId
#         self.sub = sub

class Academic:
    def __init__(self, course = None, CGPA = None):
        self.CGPA = CGPA
        self.course = course

class Sports:
    def __init__(self, sport_name = None, level = None):
        self.sport_name = sport_name
        self.level = level

class AllRounderStudent(Student, Academic, Sports):
    def __init__(self, name, age, sId, rollNum, is_academic = False, is_sports = False, course = None, CGPA = None, sport_name = None, level = None):
        super().__init__(name, age, sId, rollNum)
        self.is_academic = is_academic 
        self.is_sports = is_sports

        if self.is_academic:
            Academic.__init__(self, course, CGPA)
        
        if self.is_sports:
            Sports.__init__(self, sport_name, level)

    def display(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Roll Num: {self.rollNum}")

        if self.is_academic:
            print(f"Course: {self.course}")
            print(f"CGPA: {self.CGPA}")
        
        if self.is_sports:
            print(f"Sport Name: {self.sport_name}")
            print(f"Sport Level: {self.level}")



s1 = AllRounderStudent("Kshitij",25,102,20,is_academic=True,course="IT",CGPA=8.02)
s1.display()
```
