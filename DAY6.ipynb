```python
# Access Specifiers
# Public 
class A:
    x = 10
    def display(self):
     print(self.x)

obj = A()
obj.display()
print(obj.x)
```

```python
# Private 
# uses  __var/method name

# Protected
# uses _ var/method name

class Temp:
    __var = 10

    # get and set method used for private specifiers

    def get(self):
        print(self.__var)
    def set(self, new_val):
        self.__var = new_val

obj = Temp()
obj.get()
obj.set(20)
obj.get()
```

# ABSTRACTION

* ABC -> Abstract base class, inherit from this to make it abstract classs
* @abstractmethod -> for making method abstract
* abstract methods are meant to be inherited
* child class should implement every method from abstract class
* abstract class objects cannot be created, it works as a template
* class which inehrits the abstract class and implements all methods is called conrete class


```python
from abc import ABC, abstractmethod

class Bank(ABC):                    # Abstract Class
    @abstractmethod
    def rate_of_interest(self):     # Abstract Method
        pass

class SBI(Bank):
    def rate_of_interest(self):
        print("ROI is 7%")          # Concrete Class


# obj1 = Bank()
#cannot create object of abstract class

obj2 = SBI()
obj2.rate_of_interest()
```

```python
from abc import ABC, abstractmethod

class BankAcc(ABC):
    def __init__(self, cust_name, __date_of_birth):
        self.cust_name = cust_name
        # def setdob(__date_of_birth):
        #     self.__date_of_birth = __date_of_birth
        self.__date_of_birth = __date_of_birth
    
    @abstractmethod
    def deposit(amount):
        pass

    @abstractmethod
    def withdraw(amount):
        pass


class SavingsAcc(BankAcc):
    def __init__(self, cust_name, __date_of_birth, accNum):
        super().__init__(cust_name, __date_of_birth)
        self.__accNum = accNum

    min_bal = 500
    __curr_bal = min_bal

    def get_bal(self):
        print(self.__curr_bal)

    def display_acc_details(self):
        print(f"Name: {self.cust_name}")
        print(f"Date of Birth: {self.__date_of_birth}")
        print(f"Account Number: {self.__accNum}")

    def deposit(self, amount):
        if amount > 0 : self.__curr_bal += amount
    def withdraw(self, amount):
        if self.__curr_bal - amount >= self.min_bal: 
            self.__curr_bal -= amount
        else: print("Can withdraw above minimum balance only")

class CurrentAcc(BankAcc):

    def __init__(self, cust_name, __date_of_birth, accNum):
        super().__init__(cust_name, __date_of_birth)
        self.accNum = accNum

    overdraft_limit = 5000
    __curr_bal = 0

    def get_bal(self):
        print(self.__curr_bal)

    def display_acc_details(self):
        print(f"Name: {self.cust_name}")
        print(f"Date of Birth: {self.date_of_birth}")
        print(f"Account Number: {self.accNum}")

    
    def deposit(self, amount):
        if amount > 0: self.__curr_bal+= amount
    def withdraw(self, amount):
        if self.__curr_bal + self.overdraft_limit > amount:
            if self.__curr_bal >= amount: self.__curr_bal -= amount
            else: 
                self.overdraft_limit = self.__curr_bal + self.overdraft_limit - amount
                self.__curr_bal = 0
        else:
            print("Balance is greater than overdraft limit.")


# p1 = SavingsAcc("Kshitij", "01-01-2000", 1234)
# # p1.get_bal()
# p1.deposit(200)
# p1.get_bal()
# p1.withdraw(300)
# p1.get_bal()

p2 = CurrentAcc("Kshitj", "02-05-2003", 9874)
p2.get_bal()
p2.deposit(2000)
p2.get_bal()
p2.withdraw(-1000)
p2.get_bal()


            
```
