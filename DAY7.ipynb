```python
#COunt number of words in a file
with open('data.txt') as f:
    words = (f.read().split())
    print(len(words))
```

```python
# Extract unique words and count them
with open('data.txt') as f:
    words = f.read().split()
    unique_words = set(words)
    print(unique_words)
    print(len(unique_words))
```

```python
# Occurence of each word 
word_count = {}
with open('data.txt') as f:
    words = f.read().split()
    for i in words:
        if i in word_count:
            word_count[i] += 1
        else:
            word_count[i] = 1
    print(word_count)
```

## JSON

```python
data = {
    "Name" : "Kshitj",
    "Age" : 22,
    "Course" : "IT"
}
```

```python
import json
with open('json1.json', 'r+') as f:
    json.dump(data, f)
    # json.load(f)
```

```python
with open('json1.json', 'r+') as f:
    # json.dump(data, f)
    print(json.load(f))
```

## CSV

```python
import csv
with open('data.csv', 'w', newline= "") as file:
    writer = csv.writer(file)
    writer.writerow(["Name","Age","Course"])
    writer.writerow(["KShitij",22,"IT"])
```

```python
rows = [
    ["Name","Age"],
    ["Kshitij",22],
    ["Abhay",23]
]
with open("data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)
```

```python
import csv 
with open("data.csv","w", newline='') as file:
    fieldnames=["Name","Age"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({"Name":"Kshitij", "Age": 22})
    writer.writerow({"Name":"Aman","Age": 22})
```

```python
import csv
with open("data.csv", "r") as file:
    reader = csv.DictReader(file)
    print(reader) # -> object address
    for row in reader:
        print(row)
```
