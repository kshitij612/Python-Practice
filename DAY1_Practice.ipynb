```python
ls = [1,2,3,1,2]
if (ls[0:2] == ls[-1:-3]):
    print("Valid Slice")
```

```python
str = "madam"
if (str == str[::-1]):
    print("Palindrome")
```

```python
str = "l5sdkdkj4"

if(str[:-2:-1].isdigit()):
    print("Ends with digit.")
else:
    print("Ends with character/ special character")
```

```python
str = "kljskd"
if(str[:1].isupper()):
    print("Starts with uppercase.")
elif(str[:-2:-1].islower()):
    print("Ends with lower case")
else:
    print("Other Case")
```

```python
ls = [100,2,100,2]
if(sum(ls[:len(ls)//2]) > sum(ls[:len(ls)//2-1:-1])):
    print("First Half Greater")
else:
    print("Second Half Greater")
```
